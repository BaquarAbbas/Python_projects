```python
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
marvel = pd.read_csv('charcters_stats.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment'].value_counts()
```

```python
good = marvel[marvel['Alignment'] == 'good'] 
```

```python
good.head()
```

```python
good.sort_values(by = ['Speed'],ascending = False).head(10)
```

```python
max_power=good.sort_values(by = ['Power'],ascending = False)
max_power.head(5)
```

```python
max_speed = good[good['Speed'] == 100]
max_speed
```

```python
low_speed = good[good['Speed'] == 1]
low_speed.head(5)
```

```python
powerfull_character = good[good['Power'] == 100]
powerfull_character
```

```python
powerfull_character.shape
```

```python
good_character = good.sort_values(by = ['Total'],ascending = False)
good_character.head(10)
```

```python
good_character.shape
```

```python
#Making a barplot based on total values 
plt.figure(figsize = (11,8)) 
plt.bar(list(good_character['Name'])[0:5],list(good_character['Total'])[0:5],
        color = ['green','darkblue','silver','red','blue'])
plt.show()
```

```python
#Villians 
bad = marvel[marvel['Alignment'] == 'bad']
bad.head(6)
```

```python
#sorting values by speed 
bad.sort_values(by = ['Speed'],ascending = False).head()
```

```python
#sorting values based on Intelligence 
bad.sort_values(by = ['Intelligence'],ascending = False).head()
```

```python
powerfull_bad_character = bad.sort_values(by = ['Total'],ascending = False)
powerfull_bad_character.shape
```

```python
powerfull_bad_character.head(6)
```

```python
#Making a barplot based on name and totalvalues for villians 
plt.figure(figsize = (11,8))
plt.bar(list(powerfull_bad_character['Name'])[0:5],list(powerfull_bad_character['Total'][0:5]),
        color = ['black','grey','yellow','green','silver'])
plt.show()
```

```python
#making a histogram for good superheroes based on speed values 
plt.figure(figsize=(10,6))
plt.hist(good['Speed'],bins = 10,color = "red") 
plt.title("Distribution of speed",fontsize = 10)
plt.xlabel('Speed',fontsize = 12)
plt.ylabel("Number of good characters",fontsize = 12)
plt.show()
```

```python
#making a histogram for supervillians based on combatskills 
plt.figure(figsize = (10,6))
plt.hist(bad['Combat'])
plt.title('Combat skills of various supervillians',fontsize = 15)
plt.xlabel('Combat',fontsize = 12)
plt.ylabel('Number of bad characters',fontsize = 12)
plt.show()
```
